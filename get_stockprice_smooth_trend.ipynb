```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('D:/College/ALA/Assignment 1/GOOG.csv')  


closing_prices = data['close']


window_size = 30  
smoothed_prices = np.convolve(closing_prices, np.ones(window_size)/window_size, mode='valid')


plt.figure(figsize=(10, 8))
plt.plot(closing_prices,label='Original Closing Prices')
plt.plot(smoothed_prices, label='Smoothed Closing Prices')
plt.xlabel('Days')
plt.ylabel('Closing Price')
plt.title('Google Stock Price Trend (Smoothed)')
plt.legend()
plt.show()
```
